```python
import pandas as pd
import sqlite3
import os
from pathlib import Path
```

```python
data_path = Path('../location_pulled_data')

def csvs_to_sql(csvs):
    db_name = os.path.basename(csvs) + '.db'
    dbconn = sqlite3.connect(db_name)
    files = csvs.glob('*.csv')
    
    for f in files:
        df = pd.read_csv(f)
        df.to_sql(db_name, dbconn, if_exists='append', index=False)
        
    dbconn.close()

def find_dirs():
    dirs = []
    
    for subdir in data_path.iterdir():
        if subdir.is_dir():
            dirs.append(subdir)
            
    return dirs

dirs = find_dirs()

for dir in dirs:
    csvs_to_sql(dir)
```
